```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
```

```python
plt.style.use('seaborn-ticks')
```

```python
x = np.linspace(-3*np.pi, 3*np.pi, 128)
s, c = np.sin(x), np.cos(x)
y = np.cos(x) * np.exp(-x/8)
```

```python
f, a = plt.subplots(2, figsize=(11, 10), dpi=150)
a[0].plot(x, c, ':r', label='cos(x)')
a[0].plot(x, s, 'g-',label='sin(x)')
a[0].plot(x, s**2, 'p-',label='sin(x$^2$)', ms=2)
a[0].plot(x, np.sin(2*x), 'c:s', label=u'sin(2x)汉语也行', ms=2)
a[0].plot(x, c*2, 'k-s', ms=1, label='cos(x)$\\times$2')
a[0].plot(x, y, 'b^', label='cos(x)$e^{-x/8}$', ms=2)
a[0].plot(x, x<-1, 'g--', label='Sharp Interface')
a[0].legend()
```

```python
xl = np.linspace(-1, 1, 9)
```

```python
def pilabel(x):
  if x == 1:
    return f'$\pi$'
  elif x == -1:
    return f'-$\pi$'
  elif x == 0:
    return '0'
  else:
    return f'{x:g}$\pi$'
```

```python
a[0].axis([-np.pi, np.pi, -2.1, 2.1])

a[0].set(xticks=xl*np.pi, xticklabels=list(map(pilabel, xl)))
f
```

```python
a[0].set_xticklabels(a[0].get_xticklabels(), fontsize=10, color='red', alpha=.5)
a[0].set_yticklabels(a[0].get_yticklabels(), fontsize=10, color='green', rotation=45)
a[0].legend(loc=(0.56,0))
a[0].set_title('Sine & Cosine', size=30, color='blue', style='italic', weight='bold')
a[0].set_xlabel('X$\degree$', size=15, color='pink')
a[0].set_ylabel('Values', size=15, color='purple')
a[0].annotate('Local Max', xy=(0,2), xytext=(1.5,1.5), arrowprops={}, color='red', size=15)
f
```

```python
aa = f.add_axes([.256, 0.55, .3, .06])
aa.xaxis.set_visible(0)
aa.yaxis.set_visible(0)
aa.plot(x, c, ':r', label='cos(x)')
aa.plot(x, s, 'g-',label='sin(x)')
aa.plot(x, s**2, 'p-',label='sin(x$^2$)', ms=1)
aa.plot(x, np.sin(2*x), 'c:s', label='sin(2$\\times$x)', ms=1)
aa.plot(x, c*2, 'k-s', ms=1, label='cos(x)$\\times$2', lw=1)
aa.plot(x, y, 'b^', label='complicated', ms=1)
aa.plot(x, x<0, 'g--', label='Sharp Interface', lw=1)
aa.add_patch(matplotlib.patches.Rectangle((-np.pi, -2), 2*np.pi, 4, facecolor='cyan', alpha=.5))
aa.annotate('index', xy=(0,0), xytext=(0,3.2))
a[0].text(-.10,1.5,'text', fontsize = 20, bbox={'alpha':.5,'facecolor':'pink'})
plt.tight_layout()
f
```
