# Initialazation
Import necessary libararies.

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
```

Pick a Style Sheet or use the default one. You can even make your own. I personnally don't think I will do that. Great jobs have been done already.

```python
print(plt.style.available)
plt.style.use('seaborn-ticks')
```

# Data preparation
the data we are going to use in this guide

```python
x = np.linspace(-3*np.pi, 3*np.pi, 128)
s, c = np.sin(x), np.cos(x)
y = np.cos(x) * np.exp(-x/8)
```

# Plotting
Real job of plotting

Creat the figure and axeses with subplots()

```python
f, a = plt.subplots(2, figsize=(11, 10))
```

Add plottings to the first axes

```python
a[0].plot(x, c, ':r', label='cos(x)')
a[0].plot(x, s, 'g-',label='sin(x)')
a[0].plot(x, s**2, 'p-',label='sin(x$^2$)', ms=3)
a[0].plot(x, np.sin(2*x), 'c:s', label=u'sin(2x)汉语', ms=2)
a[0].plot(x, c*2, 'k-s', ms=1, label='cos(x)$\\times$2')
a[0].plot(x, y, 'b^', label='cos(x)$e^{-x/8}$', ms=2)
f
```

Add legend

```python
a[0].legend()
f
```

Set the axises limit. There are other ways to do this. I found this the most convenient for me.

```python
a[0].axis([-np.pi, np.pi, -2.1, 2.1])
f
```

Relocate the legend, there are more flexible way. But, to me this is good enough

```python
a[0].legend(loc=(0.56,0))
f
```

Add something to the second axes

```python
a[1].plot(x, x**2, 'o', label='$x^2$', ms=2)
a[1].plot(x, x**3, '--', label='$x^3$')
a[1].plot(x, x*10, 'r', label='10x')
a[1].axis([-2.5, 2.5,-10, 10])
a[1].legend(loc=0)
f
```

Helper function to help format x axis, the result will show on the graph

```python
xl = np.linspace(-1, 1, 9)
def pilabel(x):
  if x == 1:
    return f'$\pi$'
  elif x == -1:
    return f'-$\pi$'
  elif x == 0:
    return '0'
  else:
    return f'{x:g}$\pi$'
```

Make x axis ticks at muliple of $\pi$, and ticks lable as that with the helper function

```python
a[0].set(xticks=xl*np.pi, xticklabels=list(map(pilabel, xl)))
f
```

Set the tick labels size, color and rotation. There are other ways to do that, this is the most conenient way for me. And it is the most staightforward way when you understand it.

```python
a[0].set_xticklabels(a[0].get_xticklabels(), fontsize=15, color='red' , weight='bold')
a[0].xaxis.set_tick_params(labeltop=True, tick2On=True)
a[0].set_yticklabels(a[0].get_yticklabels(), fontsize=10, color='green', rotation=45, verticalalignment	='top')
f
```

Set the title and axis labels(size, color, rotation, style et. al.)

```python
a[0].set_title('Sine & Cosine', size=30, color='blue', style='italic', weight='bold',pad=45)
a[0].set_xlabel('X$\degree$', size=15, color='red', alpha=.5, weight='bold')
a[0].set_ylabel('Values', size=15, color='purple')
f
```

Add an annotation to the graph

```python
a[0].annotate('Local Max', xy=(0,2), xytext=(1.5,1.5), arrowprops={}, color='red', size=15)
f
```

Add a Text box

```python
a[0].text(-.10,1.5,'text', fontsize = 20, bbox={'alpha':.5,'facecolor':'pink'})
f
```

Add an extra axes as an index

```python
aa = f.add_axes([.256, 0.55, .3, .06])
f
```

Remove the index Ticks and Label

```python
aa.xaxis.set_visible(0)
aa.yaxis.set_visible(0)
f
```

Fill the index with the original plot

```python
aa.plot(x, c, ':r', label='cos(x)')
aa.plot(x, s, 'g-',label='sin(x)')
aa.plot(x, s**2, 'p-',label='sin(x$^2$)', ms=1)
aa.plot(x, np.sin(2*x), 'c:s', label='sin(2$\\times$x)', ms=1)
aa.plot(x, c*2, 'k-s', ms=1, label='cos(x)$\\times$2', lw=1)
aa.plot(x, y, 'b^', label='complicated', ms=1)
f
```

Add a box to show the position of the main graph on the index

```python
aa.add_patch(matplotlib.patches.Rectangle((-np.pi, -2.1), 2*np.pi, 4.2, facecolor='cyan', alpha=.5))
f
```

Add a title to the index


```python
aa.set_title('INDEX', fontsize=10, pad=1)
f
```

Tight layout if necessary

```python
plt.tight_layout()
f
```
